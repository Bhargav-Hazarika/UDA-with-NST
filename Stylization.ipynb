```python
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import os
from tqdm import tqdm
```

```python
def calc_mean_std(feat, eps=1e-5):
    size = feat.size()
    N, C = size[:2]
    feat_var = feat.view(N, C, -1).var(dim=2) + eps
    feat_std = feat_var.sqrt().view(N, C, 1, 1)
    feat_mean = feat.view(N, C, -1).mean(dim=2).view(N, C, 1, 1)
    return feat_mean, feat_std

def adain(content_feat, style_feat):
    c_mean, c_std = calc_mean_std(content_feat)
    s_mean, s_std = calc_mean_std(style_feat)
    normalized = (content_feat - c_mean) / c_std
    return normalized * s_std + s_mean
```

```python
decoder = nn.Sequential(
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 256, (3, 3)),
    nn.ReLU(),
    nn.Upsample(scale_factor=2, mode='nearest'),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 256, (3, 3)),
    nn.ReLU(),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 256, (3, 3)),
    nn.ReLU(),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 256, (3, 3)),
    nn.ReLU(),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 128, (3, 3)),
    nn.ReLU(),
    nn.Upsample(scale_factor=2, mode='nearest'),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(128, 128, (3, 3)),
    nn.ReLU(),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(128, 64, (3, 3)),
    nn.ReLU(),
    nn.Upsample(scale_factor=2, mode='nearest'),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(64, 64, (3, 3)),
    nn.ReLU(),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(64, 3, (3, 3)),
)

vgg = nn.Sequential(
    nn.Conv2d(3, 3, (1, 1)),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(3, 64, (3, 3)),
    nn.ReLU(),  # relu1-1
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(64, 64, (3, 3)),
    nn.ReLU(),  # relu1-2
    nn.MaxPool2d((2, 2), (2, 2), (0, 0), ceil_mode=True),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(64, 128, (3, 3)),
    nn.ReLU(),  # relu2-1
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(128, 128, (3, 3)),
    nn.ReLU(),  # relu2-2
    nn.MaxPool2d((2, 2), (2, 2), (0, 0), ceil_mode=True),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(128, 256, (3, 3)),
    nn.ReLU(),  # relu3-1
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 256, (3, 3)),
    nn.ReLU(),  # relu3-2
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 256, (3, 3)),
    nn.ReLU(),  # relu3-3
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 256, (3, 3)),
    nn.ReLU(),  # relu3-4
    nn.MaxPool2d((2, 2), (2, 2), (0, 0), ceil_mode=True),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(256, 512, (3, 3)),
    nn.ReLU(),  # relu4-1, this is the last layer used
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU(),  # relu4-2
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU(),  # relu4-3
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU(),  # relu4-4
    nn.MaxPool2d((2, 2), (2, 2), (0, 0), ceil_mode=True),
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU(),  # relu5-1
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU(),  # relu5-2
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU(),  # relu5-3
    nn.ReflectionPad2d((1, 1, 1, 1)),
    nn.Conv2d(512, 512, (3, 3)),
    nn.ReLU()  # relu5-4
)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
encoder = encoder.to(device).eval()
decoder = decoder.to(device).eval()
```

### Image Transform

```python
transform = transforms.Compose([
    transforms.Resize(512),
    transforms.CenterCrop(512),
    transforms.ToTensor()
])

def load_image(path):
    img = Image.open(path).convert("RGB")
    return transform(img).unsqueeze(0).to(device)
```

```python
@torch.no_grad()
def stylize(content_path, style_path, alpha=1.0):
    content = load_image(content_path)
    style = load_image(style_path)

    content_feat = encoder(content)
    style_feat = encoder(style)

    t = adain(content_feat, style_feat)
    t = alpha * t + (1 - alpha) * content_feat

    out = decoder(t)
    return out.clamp(0, 1)
```

```python
def collect_image_files(root, exts={".jpg", ".jpeg", ".png"}):
    files = []
    for dirpath, _, filenames in os.walk(root):
        for fn in filenames:
            if os.path.splitext(fn)[1].lower() in exts:
                files.append(os.path.join(dirpath, fn))
    return sorted(files)
```

```python
STYLIZED_ROOT = GDRIVE_BASE + "/stylized"
MANIFEST_OUT  = GDRIVE_BASE + "/manifests/stylized_manifest.csv"
CONTENT_DIR   = GDRIVE_BASE + "/preproc/images_jpeg"
STYLE_DIR     = GDRIVE_BASE + "/styles"

K = 3
ALPHA = 1.0

os.makedirs(STYLIZED_ROOT, exist_ok=True)
```

```python
import pandas as pd
```

```python
# -----------------------------
# Load or create stylized manifest
# -----------------------------
if os.path.exists(MANIFEST_OUT):
    stylized_df = pd.read_csv(MANIFEST_OUT)
else:
    stylized_df = pd.DataFrame(columns=[
        "content_id",
        "content_path",
        "style_group",
        "style_image",
        "variant_id",
        "stylized_path",
        "alpha"
    ])

done_keys = set(
    zip(
        stylized_df.content_id,
        stylized_df.style_group,
        stylized_df.variant_id
    )
)
```

```python
# -----------------------------
# Collect content images
# -----------------------------
# content_files = sorted([
#     f for f in os.listdir(CONTENT_DIR)
#     if f.lower().endswith(".jpg")
# ])

SAMPLE_CSV = GDRIVE_BASE + "/manifests/sampling_small.csv"

subset_df = pd.read_csv(SAMPLE_CSV)

# assuming 'id' column exists and matches filenames like 000123.jpg
content_files = sorted([
    f"{int(row['id']):06d}.jpg"
    for _, row in subset_df.iterrows()
])
```

### Sanity Check

```python
print("Total sampled images:", len(content_files))
print("First 5:", content_files[:5])

# Verify files actually exist
missing = [
    f for f in content_files
    if not os.path.exists(os.path.join(CONTENT_DIR, f))
]
print("Missing files:", len(missing))
```

```python
# -----------------------------
# Collect style images per group
# -----------------------------
style_groups = {}
for sg in sorted(os.listdir(STYLE_DIR)):
    sg_path = os.path.join(STYLE_DIR, sg)
    if os.path.isdir(sg_path):
        imgs = collect_image_files(sg_path)
        if len(imgs) >= K:
            style_groups[sg] = imgs

assert len(style_groups) > 0, "No valid style groups found."
```

```python
# -----------------------------
# Stylization loop
# -----------------------------
rows = []

for cf in tqdm(content_files, desc="AdaIN Precompute"):
    content_id = cf.replace(".jpg", "")
    content_path = os.path.join(CONTENT_DIR, cf)

    base_idx = int(content_id)

    for style_group, style_imgs in style_groups.items():
        for v in range(K):
            key = (content_id, style_group, v)
            if key in done_keys:
                continue  # resume-safe

            style_img = style_imgs[(base_idx + v) % len(style_imgs)]

            # safety: avoid identical image
            if os.path.samefile(content_path, style_img):
                continue

            out = stylize(content_path, style_img, alpha=ALPHA)

            out_dir = os.path.join(STYLIZED_ROOT, style_group)
            os.makedirs(out_dir, exist_ok=True)

            out_name = f"{content_id}_v{v}.jpg"
            out_path = os.path.join(out_dir, out_name)

            transforms.ToPILImage()(out.squeeze(0)).save(out_path)

            rows.append({
                "content_id": content_id,
                "content_path": content_path,
                "style_group": style_group,
                "style_image": style_img,
                "variant_id": v,
                "stylized_path": out_path,
                "alpha": ALPHA
            })


```

```python
# -----------------------------
# Save manifest incrementally
# -----------------------------
if rows:
    stylized_df = pd.concat([stylized_df, pd.DataFrame(rows)], ignore_index=True)
    stylized_df.to_csv(MANIFEST_OUT, index=False)

print("Stylization complete. Total stylized images:", len(stylized_df))
```
